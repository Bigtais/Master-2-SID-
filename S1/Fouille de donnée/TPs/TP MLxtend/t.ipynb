```python
from scipy.stats import shapiro

# Données pour chaque condition
data_control = [62508, 54328, 43417, 32782, 55050, 39322]
data_high_K = [58855, 35281, 48645, 62071, 48177, 56713]
data_NMDA_antagonist = [32681, 24755, 21321, 28633, 17138, 21787]

# Test de normalité de Shapiro-Wilk
stat_control, p_value_control = shapiro(data_control)
stat_high_K, p_value_high_K = shapiro(data_high_K)
stat_NMDA_antagonist, p_value_NMDA_antagonist = shapiro(data_NMDA_antagonist)

# Affichage des résultats
print("Test Shapiro-Wilk pour le contrôle : Statistique =", stat_control, ", p-valeur =", p_value_control)
print("Test Shapiro-Wilk pour le haut K+ : Statistique =", stat_high_K, ", p-valeur =", p_value_high_K)
print("Test Shapiro-Wilk pour l'antagoniste NMDA : Statistique =", stat_NMDA_antagonist, ", p-valeur =", p_value_NMDA_antagonist)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Fonction pour générer le graphe Q-Q
def plot_qq(data, title):
    plt.figure(figsize=(8, 6))
    sm.qqplot(np.array(data), line='s')
    plt.title(title)
    plt.show()

# Génération des graphes Q-Q pour chaque condition
plot_qq(data_control, "Q-Q Plot - Contrôle")
plot_qq(data_high_K, "Q-Q Plot - Haut K+")
plot_qq(data_NMDA_antagonist, "Q-Q Plot - Antagoniste NMDA")
```

```python
from scipy.stats import f_oneway

# Données pour chaque condition
A1 = [62508, 54328, 43417, 32782, 55050, 39322]
A2 = [58855, 35281, 48645, 62071, 48177, 56713]
A3 = [32681, 24755, 21321, 28633, 17138, 21787]

# Test de One-Way ANOVA
statistic, p_value = f_oneway(A1, A2, A3)

# Affichage des résultats
print("Statistique de test de One-Way ANOVA =", statistic)
print("P-valeur =", p_value)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Données pour chaque condition
data_A1 = [62508, 54328, 43417, 32782, 55050, 39322]
data_A2 = [58855, 35281, 48645, 62071, 48177, 56713]
data_A3 = [32681, 24755, 21321, 28633, 17138, 21787]

# Création d'un DataFrame pour utiliser Seaborn
import pandas as pd

df = pd.DataFrame({
    'Condition': ['A1']*len(data_A1) + ['A2']*len(data_A2) + ['A3']*len(data_A3),
    'Values': data_A1 + data_A2 + data_A3
})

# Boîtes à moustaches
plt.figure(figsize=(10, 6))
sns.boxplot(x='Condition', y='Values', data=df)
plt.title('Boîtes à moustaches pour l\'ANOVA univariée')
plt.xlabel('Condition')
plt.ylabel('Valeurs')

# Affichage du graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Données pour chaque condition
data_A1 = [62508, 54328, 43417, 32782, 55050, 39322]
data_A2 = [58855, 35281, 48645, 62071, 48177, 56713]
data_A3 = [32681, 24755, 21321, 28633, 17138, 21787]

# Création d'un DataFrame pour utiliser Seaborn
import pandas as pd

df = pd.DataFrame({
    'Condition': ['A1']*len(data_A1) + ['A2']*len(data_A2) + ['A3']*len(data_A3),
    'Values': data_A1 + data_A2 + data_A3
})

# Boîtes à moustaches avec des points
plt.figure(figsize=(10, 6))
sns.boxplot(x='Condition', y='Values', data=df, showfliers=False)  # Masquer les valeurs aberrantes
sns.swarmplot(x='Condition', y='Values', data=df, color='black', alpha=0.7)  # Ajouter des points

plt.title('Boîtes à moustaches avec points pour l\'ANOVA univariée')
plt.xlabel('Condition')
plt.ylabel('Valeurs')

# Affichage du graphique
plt.show()
```
